```python
plt.plot(m_1)
```

```python

peaks, properties = find_peaks(m_1, height=0.02)

print(f'Index of each peaks : {peaks}')
print(f'Heigth of each peaks : {properties["peak_heights"]}')

plt.plot(m_1)
plt.plot(peaks, m_1[peaks], "X")

```

```python
MAX1 = properties["peak_heights"]
print('MAX1:',MAX1)

t2 =  6.5
p = 0.37
sigma = np.std(MAX1)
mean = np.mean(MAX1)
gamma = 0.577
beta = (np.sqrt(6)/np.pi)*sigma
mu = mean - ((np.sqrt(6)/np.pi)*sigma*gamma)
l = -np.log(p)
MPM1 = -np.log(l)*beta + mu
print('MPM1:',MPM1)
```
