```python
import numpy as np
```

```python
from scipy.signal import find_peaks
```

**1.6MESH**

```python
plt.plot(m2_16)
```

```python
peaks, properties = find_peaks(m2_16, height=0.02)

print(f'Index of each peaks : {peaks}')
print(f'Heigth of each peaks : {properties["peak_heights"]}')

plt.plot(m2_16)
plt.plot(peaks, m2_16[peaks], "X")
```

```python
MAX = properties["peak_heights"]
MAX
```

```python
sorted_index_array = np.argsort(MAX)
sorted_array = MAX[sorted_index_array]

n=20
MAX_rslt = sorted_array[-n : ]
MAX_rslt
```

```python
t2 =  6.5
p = 0.37
sigma = np.std(MAX_rslt)
mean = np.mean(MAX_rslt)
gamma = 0.577
```

```python
beta = (np.sqrt(6)/np.pi)*sigma
mu = mean - ((np.sqrt(6)/np.pi)*sigma*gamma)
```

```python
l = -np.log(p)
```

```python
MPM = -np.log(l)*beta + mu
MPM
```

```python
MPM_SUCC = MPM - 8.2223
MPM_SUCC
```
