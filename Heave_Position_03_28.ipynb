**데이터 불러오기**

```python
from google.colab import files
myfile = files.upload()
```

```python
import io
import pandas as pd
```

```python
mesh1_6 = pd.read_csv(io.BytesIO(myfile['1.6mesh_position.csv']))
mesh2_85 = pd.read_csv(io.BytesIO(myfile['2.85mesh_position.csv']))
mesh4_5 = pd.read_csv(io.BytesIO(myfile['4.5mesh_position.csv']))
mesh6 = pd.read_csv(io.BytesIO(myfile['6mesh_position.csv']))
```

```python
print(mesh1_6)
print(mesh2_85)
print(mesh4_5)
print(mesh6)
```

```python

t_16 = mesh1_6['t']
m_16 = mesh1_6['m']

t_285 = mesh2_85['t']
m_285 = mesh2_85['m']

t_45 = mesh4_5['t']
m_45 = mesh4_5['m']

t_6 = mesh6['t']
m_6 = mesh6['m']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t_16, m_16, label='mesh_1.6')
plt.plot(t_285, m_285, label='mesh_2.85')
plt.plot(t_45, m_45, label='mesh_4.5')
plt.plot(t_6, m_6, label='mesh_6')



plt.xlabel('Time(s)')
plt.ylabel('Position')
plt.legend(loc='lower right')


plt.show()
```
